```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Define the numerical columns to be normalized
cols = ['age', 'salary', 'experience']

# Min-Max Normalization
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[cols] = scaler_minmax.fit_transform(df[cols])

# Z-score Normalization
scaler_standard = StandardScaler()
df_standard = df.copy()
df_standard[cols] = scaler_standard.fit_transform(df[cols])

# Plot Bar Graphs for Original and Normalized Data
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Original Data
df[cols].mean().plot(kind='bar', ax=axes[0], color=['blue', 'green', 'red'])
axes[0].set_title("Original Data (Mean Values)")

# Min-Max Normalized Data
df_minmax[cols].mean().plot(kind='bar', ax=axes[1], color=['blue', 'green', 'red'])
axes[1].set_title("Min-Max Normalized Data (Mean Values)")

# Z-score Normalized Data
df_standard[cols].mean().plot(kind='bar', ax=axes[2], color=['blue', 'green', 'red'])
axes[2].set_title("Z-score Normalized Data (Mean Values)")

plt.tight_layout()
plt.show()
```
